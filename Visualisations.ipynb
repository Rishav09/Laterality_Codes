```python
import pandas as pd
```

```python
df_train = pd.read_csv("/Users/swastik/ophthalmology/Master_Daddy_Project/Laterality/Stage_files/Split_Final_Staged_Files/Total_Equal_Split/total_laterality_train_2023-03-30_17-42-47.csv")
df_train
```

```python
interior_left = df_train['Interior_Left'].sum()
interior_right = df_train['Interior_Right'].sum()
canon_left = df_train['Canon_Left'].sum()
canon_right = df_train['Canon_Right'].sum()
other = df_train['Other'].sum()
exterior = df_train['Exterior'].sum()
```

```python
df_sum_train = pd.DataFrame({'interior_right':[interior_right],'interior_left':[interior_left],'canon_left':[canon_left],'canon_right':[canon_right],'other':[other],'exterior':[exterior]})
df_sum_train
```

```python
import matplotlib.pyplot as plt
ax = df_sum_train.plot.bar(rot=0)
# Add labels to the bars
for p in ax.containers:
    ax.bar_label(p, label_type='edge')
plt.show()
```
