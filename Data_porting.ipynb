```python
folder_path = '/home/iscreen/data_division/Stage_Data/'
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        file_name_without_ext, _ = os.path.splitext(filename)
        new_file_path = os.path.join(folder_path, file_name_without_ext + '.jpg')
        os.rename(file_path, new_file_path)
```

```python
from collections import defaultdict

extensions_count = defaultdict(int)

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        _, file_extension = os.path.splitext(filename)
        extensions_count[file_extension.lower()] += 1

print("File extensions and their count in the folder:")
for extension, count in extensions_count.items():
    print(f"{extension}: {count}")
```
