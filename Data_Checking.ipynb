```python
import pandas as pd
master_path = '/Users/swastik/ophthalmology/Master_Daddy_Project/Laterality/Stage_files/Split_Final_Staged_Files/total_laterality_master_file.csv'
df_master_file = pd.read_csv(master_path)
df_master_file
```

```python
df_master_file['labels'].value_counts()
```

```python
interior_left = df_master_file['Interior_Left'].sum()
interior_right = df_master_file['Interior_Right'].sum()
canon_left = df_master_file['Canon_Left'].sum()
canon_right = df_master_file['Canon_Right'].sum()
other = df_master_file['Other'].sum()
exterior = df_master_file['Exterior'].sum()
```

```python
df_master_train = pd.DataFrame({'interior_right':[interior_right],'interior_left':[interior_left],'canon_left':[canon_left],'canon_right':[canon_right],'other':[other],'exterior':[exterior]})
df_master_train
```

```python
import matplotlib.pyplot as plt
ax = df_master_train.plot.bar(rot=0)
# Add labels to the bars
for p in ax.containers:
    ax.bar_label(p, label_type='edge')
plt.show()
```
